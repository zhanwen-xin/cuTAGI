```python
import scipy.io as sio
import pandas as pd
import numpy as np
```

```python
time_series_path = './data/synthetic/AR.csv'
obs = pd.read_csv(time_series_path, skiprows=1,header=None, dtype=float)[0]

import matplotlib.pyplot as plt
plt.plot(obs)
```

```python
# Define AR
phi_ar = 0.75
std_ar = 0.2
mu_AR = 0.1

# Add AR to the time series
for i in range(1, len(obs)):
    mu_AR = phi_ar * mu_AR + np.random.normal(0, std_ar)
    obs[i] += mu_AR
```
