```python
from pytagi.nn import Linear, OutputUpdater, ReLU, Sequential, EvenExp
import numpy as np
import torch
```

```python
class TAGI_Net():
    def __init__(self, n_observations, n_actions):
        super(TAGI_Net, self).__init__()
        self.net = Sequential(
                    Linear(n_observations, 1000),
                    ReLU(),
                    Linear(1000, 1000),
                    ReLU(),
                    Linear(1000, n_actions),
                    # EvenExp()
                    )
        self.n_actions = n_actions
        self.n_observations = n_observations
    def forward(self, mu_x, var_x):
        return self.net.forward(mu_x, var_x)
```

```python
test_net = TAGI_Net(1, 4)
# test_net.net.eval()
for itr in range(100):
    for state in range(16):
        state = state
        state_np = np.array([state], dtype=np.float32)
        ma, Sa = test_net.net(state_np)
        print(state)
        print(ma)
        print(Sa)
        print('=====================')
```
